```python
# calculate mean abundance in donor vs recipient 

import pandas as pd 
import numpy as np 
```

```python
input_file = "important_otus.csv"

otu = pd.read_table(input_file, sep=",", index_col=0)


donors = ["5.AV_249", "7.AV_360","1.AV_214", "3.AV_209"]
recipients = ["101.AVd_295", "30.AVd_148", "34.AVd_166", "38.AVd_292", "44.AVd_150", "48.AVd_167", "52.AVd_296", "53.AVd_298", "99.AVd_290"]
```

```python
sum_row = otu.sum()
min_sum = min(otu.sum())
normalise = otu * 100 / sum_row 

# normalise data: relative abundance - multiply each row by 100 and divide by the sum of the row
otu_filtered = normalise[(normalise > 1).any(axis=1)]
```
